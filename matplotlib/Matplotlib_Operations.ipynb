```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,5,7,9,1])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.title("Num-Plot")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], 'ro')
plt.grid()
plt.show()
```

```python
import numpy as np

a = np.arange(0, 5, 0.2)

plt.plot(a, a**2, "b--", label="^2")
plt.plot(a, a**2.5, "rs", label="^2.5")
plt.plot(a, a**3, "g^", label="^3")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], linewidth=10)
plt.grid()
plt.show()
```

```python
# Using keyword arguments:
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = x1
y2 = [2,4,6,8]

lines = plt.plot(x1, y1, x2, y2)

plt.setp(lines[0], color='r', linewidth=2)
plt.setp(lines[1], color='g', linewidth=2)

plt.grid()
plt.show()
```
